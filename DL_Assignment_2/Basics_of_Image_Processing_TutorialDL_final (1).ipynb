###### Task2.1: RGB to Gray  

```python
def RGB2Gray(image):
    '''
    This function converts an RGB image to grayscale
    image: RGB image
    '''
    #write you code here and visualize the result
    gray=0.29*image[:,:,0]+0.587*image[:,:,1]+0.114*image[:,:,2]


    return gray 
a=RGB2Gray(image)
plt.imshow(a,cmap='gray')
plt.title('gray image')
print(a.shape)
    #'gray' is grayscale image, converted from RGB image
```

######  Task2.3:  Crop a 5*5 section of a gray image

```python
def ImageCrop(image,r0,c0):
    '''
    This function crops 5*5 rectangular patch defined by image coordinates(r0,c0),(r0,c0+5),(r0+5,c0) and (r0+5,c0+5)
    of an image.
    image: Image can be RGB or gray
    r0: starting row index
    c0: starting column index
    '''
    # write code to check if input is RGB , if its RGB convert it to gray
    flag=len(image.shape)
    if flag == 3:
        grayimg=RGB2Gray(image)
        # write code to select 5*5 rectangular patch defined as above (say it 'patch')
        patch=grayimg[r0:r0+5,c0:c0+5]
        # visualize patch and print its values
    return patch
```

```python
def feature_extractor(image,kernel):
    '''
    This function performs convolution operation to a gray image. We will consider 3*3 kernel here.
    In general kernel can have shape (2n+1) * (2n+1)  where n>= 0
    image: image (can be RGB or gray); if RGB convert it to gray
    kernel: 3*3 convolution kernel
    '''
    # first convert RGB to gray if input is RGB image

    l = len(image.shape)

    if l == 3:
        #write code to convert it to gray scale
        img= RGB2Gray(image)
    

    # write code to create a zero array of size (r,c) which will store the resultant value at specific pixel locations (say it output)
    h,w = img.shape
    output= np.zeros((h,w))

    #write code to create a zero array with size (r+2,c+2) if (r,c) is the gray image size.  (say it pad_img)
    pad_img = np.zeros((h+2,w+2))


    #now copy gray image to above created array at location starting from (1,1)
    pad_img[1:-1,1:-1] = img


    #write code to convolve the image
    for row in range(h):     # use appropiate range values for row and col
        for col in range(w):
            # select 3*3 patch with center at (row,col), flatten it. flatten the kernel and take dot product between both (or directly take element wise multiplication and sum it)
            # store this scalar value to output matrix with starting location (0,0)         (alternatively one could also create a list and reshape it to output size)
            patch = pad_img[row:row+3,col:col+3]
            output[row,col] = np.sum(patch*kernel)



    return output

kernel = np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]])
conv_img = feature_extractor(image,kernel)
plt.imshow(conv_img, cmap="gray")
plt.title("Convolved image")
plt.show()
```

```python
kernel = np.array([[0,-1,0], [-1,4,-1], [0,-1,0]])  #ridge or edge detection
conv_img = feature_extractor(image,kernel)
plt.imshow(conv_img, cmap="gray")
plt.title("Convolved image with ridge kernel")
plt.show()
```

```python
kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])  #sharpen
conv_img = feature_extractor(image,kernel)
plt.imshow(conv_img, cmap="gray")
plt.title("Convolved image with sharpen kernel")
plt.show()
```

```python
kernel = (1/9) * np.array([[1,1,1], [1,1,1], [1,1,1]])  #box blur(noemalized)
conv_img = feature_extractor(image,kernel)
plt.imshow(conv_img, cmap="gray")
plt.title("Convolved image with box blur kernel")
plt.show()
```

```python
#### with predefined functions #####################
from PIL import Image
Original_Image = Image.open(r'\Users\91850\Desktop\Tutorial2_Basics_of_image_processing\Tutorial2_Baiscs_of_image_processing\lena.png')
# Rotate Image By 10 Degree in anti clock wise
rotated_image1 = Original_Image.rotate(10)
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("original Lena image")
plt.subplot(1,2,2)
plt.imshow(rotated_image1)
plt.title("Rotated Lena image")
```

```python
###witout using predifined functions ################
import math
def anti_rotate(image, angle):
    img=RGB2Gray(image)
    
    x,y=np.shape(img)

    #initializing the target image with zeros
    target_img=np.zeros((x,y))
    angle=-angle
    angle *=np.pi/180
    #finding the center 
    center_x,center_y=(x/2),(y/2)

    for xt in range(x):
        for yt in range(y):
            #finding source co-ords in terms of target 
            #xprime and yprime, a and b calculation
            xc,yc=xt-center_x,yt-center_y
            xs=np.cos(angle)*xc -np.sin(angle)*yc + center_x
            ys=np.sin(angle)*xc +np.cos(angle)*yc + center_y
            xsp,ysp=math.floor(xs),math.floor(ys)
            a,b=xs-xsp,ys-ysp
            #boundary condition for bilinear interpolation
            if xs>=0 and ys>=0 and xs<x-1 and ys<y-1:
                target_img[xt,yt]=(1-a)*(1-b)*img[xsp,ysp] + (1-a)*b*img[xsp,ysp+1] + a*(1-b)*img[xsp+1,ysp] + a*b*img[xsp+1,ysp+1]
            else:
                target_img[xt,yt]=0
    
    return target_img

a=anti_rotate(image,10)
plt.imshow(a)
plt.title("Rotated Lena image")
                


    
```
